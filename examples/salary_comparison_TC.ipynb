# 德國薪資比較範例

透過這份 Jupyter Notebook，展示如何使用 `destatis-gehaltsrechner` 來比較不同職業與地區的薪資中位數估算值。

```python
import sys
import os
# 確保可以 import 上層目錄的 package
sys.path.append(os.path.abspath('..'))

from salary_calculator import load_models, schaetze_monatsgehalt

MODELS = load_models()
```

## 1. 比較不同州的薪資（以 IT 後端工程師為例）
條件：5年年資，碩士學歷，每週 40 小時，大規模企業（>=1000人）

```python
bundeslaender = {
    'Baden-Württemberg': 'EF13_108',
    'Bayern': 'EF13_109',
    'Hessen': 'EF13_106',
    'Berlin': 'EF13_111',
    'Sachsen': 'EF13_114'
}

print("=== 薪資比較：不同州份 ===")
for state, key in bundeslaender.items():
    res = schaetze_monatsgehalt(
        berufsjahre=5,
        ausbildungsjahre=19,
        ef59u3_key="EF59U3_15",
        unternehmen_key="UNGr_UN6",
        bundesland_key=key,
        kldb_code="43414",
        vollzeit=True,
        geschlecht="gesamt",
        befristet=False,
        models=MODELS
    )
    print(f"{state:<20} : {res['median_monatsgehalt_brutto']:,.2f} €")
```

## 2. 比較不同職業別的薪資（以 Hamburg 為例）
條件：10年年資，對應學歷，每週 40 小時，中型規模企業（50-249人）

```python
berufe = [
    {"name": "軟體開發 (Softwareentwicklerin)", "kldb": "43414", "ausbildung": 17},
    {"name": "護理師 (Krankenpflege)", "kldb": "81312", "ausbildung": 13},
    {"name": "零售銷售 (Einzelhandel)", "kldb": "61212", "ausbildung": 13},
    {"name": "會計 (Buchhaltung)", "kldb": "92413", "ausbildung": 15}
]

print("\n=== 薪資比較：不同職業別 ===")
for beruf in berufe:
    res = schaetze_monatsgehalt(
        berufsjahre=10,
        ausbildungsjahre=beruf["ausbildung"],
        ef59u3_key="EF59U3_15",
        unternehmen_key="UNGr_UN3",  # 50-249 人
        bundesland_key="EF13_102",   # Hamburg
        kldb_code=beruf["kldb"],
        vollzeit=True,
        geschlecht="gesamt",
        befristet=False,
        models=MODELS
    )
    print(f"{beruf['name']:<35} : {res['median_monatsgehalt_brutto']:,.2f} €")
```
